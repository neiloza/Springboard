The user engagement file will be used to create the target variable, while the users info dataset will be used as the featureset. Null values will be filled as appropriate. let's start by encoding all users with 3 logins in 1 week as adopted

```python
user_engagement.groupby(user_engagement.user_id).agg('count').visited.value_counts()
```

```python
user_engagement.time_stamp=user_engagement.time_stamp.dt.date
```

```python
user_engagement.head()
```

```python
user_engagement.drop_duplicates()
```

```python
def adopt_func(a):
    if len(a) >= 3: 
        b = [i for i in a]  
        b.sort() 
        c = [b[i+1] - b[i] for i in range(len(b)-1)]
        d = [c[i] + c[i+1] for i in range(len(c)-1)]
        
        if (min(d) <= 7):
            return 1
        else:
            return 0
    else:
        return 0
```

```python
adopted_users=user_engagement.groupby('user_id').agg(adopt_func)
adopted_users
```

the above dataframe, adopted_users, is the target variable for this analysis. Adopted users are labeled 1 and the rest 0. some users are missing from the dataframe so we will join this table with the users table to add them in. First though, lets remove useless columns from our featureset. useless columns are things like name, email, etc. Also, I will create a new variable to represent if users were invited to the platform by another user.

```python
featureset=users.drop(['creation_time','name','email','last_session_creation_time','org_id'],axis=1)
```

```python
featureset['invited']=featureset.invited_by_user_id.isnull().apply(lambda x: 1 if x==False else 0)
featureset.head(100)
```

```python
featureset=featureset.drop('invited_by_user_id',axis=1)
```

```python
featureset=pd.get_dummies(featureset,drop_first=True)
```

```python
featureset=featureset.set_index('object_id')
featureset
```

Now that's done, lets do final preparations for the target variable and featureset before finding the most useful features

```python
adopted_users.columns=['Target']
adopted_users
```

```python
combined=featureset.join(adopted_users)
y=combined.Target.fillna(0)
y
```

I will use the random forest classifier to identify the most important features

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(featureset,y,test_size=.25)
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
```

```python
imp_feat = pd.DataFrame({'feature':featureset.columns, 'importance':rf.feature_importances_})
imp_feat.sort_values('importance', ascending=False, inplace=True)
imp_feat.head(10)
```

the people most likely to become adopted users seem to be those involved in a project of some sort, were invited, or had google authorization.
