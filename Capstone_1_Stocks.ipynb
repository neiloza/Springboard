```python
def add_new_stock(filename):
    df=pd.read_csv(filename, sep=',', index_col='Date', parse_dates=True)
    period_index=filename.find('.')
    last_slash=filename.rfind('/')
    stock_name=filename[last_slash+1:period_index]
    df['Name']=stock_name
    df=df.reset_index()
    df=df.set_index(['Name','Date'])
    global complete_stock_list
    complete_stock_list=complete_stock_list.append(df)
    
    
    
    

```
